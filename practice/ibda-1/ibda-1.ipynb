```python
years_range = np.arange(1989, 1996)
years_data  = pd.DataFrame()

for i in years_range:
    summer_temp = random.uniform(22, 33)
    winter_temp = random.uniform(-22, -12)
    years_data = pd.concat([years_data, wg.generate_weather(winter_temp, summer_temp, year=i, save=False, verbose=False, noise=False, deviation=0.5, outliers=0.05, nans=0.05)], axis=0, ignore_index=True)
years_data = years_data.dropna()
```

```python
years_data.to_csv('weather_data_1989_1996.csv', index=False)
```
